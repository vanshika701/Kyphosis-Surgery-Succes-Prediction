# Decision Tree Visualisation

```python
import pandas as pd
```

```python
from IPython.display import Image  
import six 
from sklearn.tree import export_graphviz
import pydot 
from io import StringIO
```

```python
data =pd.read_csv('Kyphosis.csv')
```

```python
data.columns
```

```python
features=data[['Age', 'Number', 'Start']]
```

```python
features
```

## Model training and Predicition

```python
X=data[['Age', 'Number', 'Start']]
y=data['Kyphosis']
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Visualization

```python
features_subset = features[:3]
dot_data = StringIO()  
export_graphviz(dtree, out_file=dot_data,feature_names=features_subset,filled=True,rounded=True)

graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png())  
```
